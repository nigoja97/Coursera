```python
def simulate_gbm(S0 = 4000, mu = 0.05, sigma = 0.18, T = 1, steps = 252, seed =1):
    np.random.seed(seed)
    dt = T/steps
    S = np.zeros(steps + 1)
    S[0] = S0
    for t in range(steps):
        Z = np.random.randn()
        S[t+1] = S[t] + mu*S[t]*dt + sigma*S[t]*np.sqrt(dt)*Z
    return S
```

```python
gbm = simulate_gbm(S0=S0, mu=mu, sigma=sigma, T=T, steps=steps, seed=1)
plt.plot(gbm)
```

```python
x0 = 0.03
theta = 0.04
kappa = 1.2
sigma = 0.25
T, steps = 2, 500
dt = T/steps

np.random.seed(0)
X = np.zeros(steps + 1)
X[0] = x0
for t in range(steps):
    Z = np.random.randn()
    drift = kappa*(theta - X[t])*dt
    diffusion = sigma*np.sqrt(max(X[t],0))*np.sqrt(dt)*Z
    X[t+1] = max(X[t] + drift + diffusion, 0)

plt.plot(X)
```
