```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2023-01-01", end="2025-11-30") 
data["Returns"] = data["Close"].pct_change()
data.to_csv("Apple_Data.csv")
```

```python
mu = data['Returns'].mean()*252
sigma = data['Returns'].std()*np.sqrt(252)  
S0 = data['Returns'].iloc[-1]
T = 1
steps = 252
dt = T/steps
np.random.seed(1)
S = np.zeros(steps + 1)
S[0] = S0
for t in range(steps):
    Z = np.random.randn()
    S[t+1] = S[t] + mu*S[t]*dt + sigma*S[t]*np.sqrt(dt)*Z   
plt.plot(S)
```
